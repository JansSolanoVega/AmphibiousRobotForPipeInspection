```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import os 
import time

PATH_FOLDER_DATASET = 'trainset/trainset/image'
VIDEO_RECORD_DATASET = '1'

VIDEO_PATH = "videoPrueba.mp4"
```

```python
def save_all_images_from_video(video_path, output_directory):
    # Open the video file
    video = cv2.VideoCapture(video_path)

    fps = video.get(1)
    # Initialize the frame counter
    frame_counter = 0
    # Check if the folder already exists
    if not os.path.exists(output_directory):
        # Create the folder
        os.makedirs(output_directory)
        print(f"Folder created at: {output_directory}")
    else:
        print("Folder already exists.")
        
    # Read and save all frames in the video
    while True:
        # Read the current frame
        ret, frame = video.read()

        # Break the loop if reading the frame was unsuccessful
        if not ret:
            break

        # Save the frame as an image
        image_path = f"{output_directory}/frame_{frame_counter}.jpg"
        cv2.imwrite(image_path, frame)
        print(f"Saved image: {image_path}")

        # Increment the frame counter
        frame_counter += 1

    # Release the video file and close any open windows
    video.release()
    cv2.destroyAllWindows()
```

```python
# Example usage
video_path = "videoPrueba5.mp4"
output_directory = os.path.join("E:/9noCiclo/PDM/LaneDetection/trainsetRNN/image", video_path.split(".")[0])

save_all_images_from_video(video_path, output_directory)
```
